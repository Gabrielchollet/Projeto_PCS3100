#A prototipagem do software foi realizada utilizando Blink e uma simulação de ambiente IoT em python, enquanto a parte do hardware foi realizada utilizando Arduino.

```python
#Simular valores de temperatura e umidade através do Blynk API
import requests
import random
import time
import csv

# Auth token do projeto
auth_token = "pMCXrYr1wV0e_hz2XlItxCZN1LepoK3E"

# Blynk Server IP
blynk_server = '45.55.96.146'

# Pin do controle
pin_controle = 'V0'

estado_controle = requests.get('http://'+blynk_server+'/'+auth_token+'/get/'+pin_controle)

# String -> Int
estado_controle = estado_controle.text.replace("[", " ").replace("]", " ").replace('"', " ").replace("'", " ").strip()

while int(estado_controle) == 1:
    # Simulação das medições de temperatura e umidade
    temp = 25 + random.randint(1, 3) - random.randint(1, 3)

    umid = 75 + random.randint(1, 5) - random.randint(1, 5)

    url_temp = "http://blynk-cloud.com/pMCXrYr1wV0e_hz2XlItxCZN1LepoK3E/update/V1?value=" + \
        str(temp)

    url_umid = "http://blynk-cloud.com/pMCXrYr1wV0e_hz2XlItxCZN1LepoK3E/update/V2?value=" + \
        str(umid)

    r_temp = requests.get(url_temp)

    r_umid = requests.get(url_umid)


    time.sleep(1)

    # Coleta os dados simulados e os armazena em um arquivo *.csv

    url_temp = "http://blynk-cloud.com/pMCXrYr1wV0e_hz2XlItxCZN1LepoK3E/get/V1"

    url_umid = "http://blynk-cloud.com/pMCXrYr1wV0e_hz2XlItxCZN1LepoK3E/get/V2"

    r_temp = requests.get(url_temp)

    r_umid = requests.get(url_umid)


    f = open('data.csv', 'a')

    with f:
        writer = csv.writer(f)

        writer.writerow([r_temp.text.replace(
            '"', '').replace('[', '').replace(']', ''), r_umid.text.replace(
            '"', '').replace('[', '').replace(']', '')])

    time.sleep(2)

    estado_controle = requests.get('http://'+blynk_server+'/'+auth_token+'/get/'+pin_controle)

    estado_controle = estado_controle.text.replace("[", " ").replace("]", " ").replace('"', " ").strip()
```
